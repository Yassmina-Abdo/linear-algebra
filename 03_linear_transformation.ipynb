```python
# Make all numpy available via shorter 'np' prefix
import numpy as np
# Make all sympy available via shorter 'sym' prefix
import sympy as sym

import math


#import plotter
import matplotlib.pyplot as plt


def rvect(m):
    return np.fix((10 * (np.random.rand(m, 1))))


def rmat(m, n):
    return np.fix((10 * (np.random.rand(m, n))))


def rref(A):
    return np.array(sym.Matrix(A).rref(iszerofunc=lambda x:sym.comp(x,0,1e-13))[0], np.float64)


def plot2d(H):
    x = H[0,:]
    y = H[1,:]
    plt.plot(x, y, 'ro', x, y, 'g-')
    plt.axis([-10, 10, -10, 10])
```

- Use the matrices A, B, C and M = CBA from this example in the following.
- Generate a random vectors
    - `x = 1000 * rvect(2)`
    - `y = A @ x`
    - `z = B @ y`
    - `w = C @ z`


The vectors x, y, z,w describe the random traffic flows through succesive parts of the network.

- Calculate:
    - `np.array([1, 1]) @ x`
    - `np.array([1, 1, 1]) @ y`
    - `np.array([1, 1, 1]) @ z`
    - `np.array([1, 1, 1, 1]) @ w`.

- Explain the meaning of each answer in terms of the traffic flow through a particular part in the network.

### ---------------------------------------------------Write your answer here--------------------------------------

1- Each answer is the Total of cars which will pass through the roads, and the total number of cars  equal to which goes out at w are the same numbers








### --------------------------------------------------------------------------------------------------------------------------

```python
A = np.array([[0.30, 0],
              [0.14, 0],
              [0.56 , 1]])
B = np.array([[0.5, 0, 0],
              [0.5, 1, 0.6],
              [0 , 0, 0.4]])
C = np.array([[1, 0.30, 0],
              [0, 0.20 , 0],
              [0 , 0.35, 0.7],
              [0 , 0.15 , 0.30]])
M= C @ B @ A
x = 1000 * rvect(2)
y = A @ x
z = B @ y
w = C @ z
print(f"M = \n{M}")
print("-----")
print(np.array([1, 1]) @ x)
print("-----")
print(np.array([1, 1, 1]) @ y)
print("-----")
print(np.array([1, 1, 1]) @ z)
print("------")
print(np.array([1, 1, 1, 1]) @ w)
```
